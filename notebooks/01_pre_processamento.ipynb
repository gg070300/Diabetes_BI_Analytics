## 2.3.1. Descrição de cada variavel

1. diabetes: variavel resposta (0 = sem diabetes, 1 = pre diabetes, 2= diabetes)
2. pressao_alta: se o paciente tem pressão alta (0 = não, 1= sim)
3. colesterol_alto: se o paciente tem colesterol alto (0 = não, 1= sim)
4. imc: indioce de massa corporal (valor numerico)
5. fumante: se o paciente fumou pelo menos 100 cigarros ao longo da vida (0 = não, 1= sim)
6. ataque_cardiaco: se o paciente ja teve infarto (0 = não, 1= sim)
7. atividade_fisica: se o paciente realizou atividade fisica nos ultimos 30 dias (0 = não, 1= sim)
8. frutas: se o paciente consome frutas pelo menos uma vez por dia (0 = não, 1= sim)
9. vegetais: se o paciente consome vegetatis pelo menos uma vez poe dia (0 = não, 1= sim)
10. alto_consumo_alcool: se o paciente tem uso excessivo de álcool (0 = não, 1= sim)
11. convenio_medico: se o paciente possui convenio medico (0 = não, 1= sim)
12. dificuldade_subir_escadas: se o paciente tem dificuldade em subir escadas (0 = não, 1= sim)
13. sexo: (0 = feminino, 1 = masculino)
14. idade: em formato de categoria: (1 = 18–24 anos, 2 = 25–29 anos, 3 = 30-34 anos, 4 = 35-39 anos, 5 = 40-44 anos, 6 = 45-49 anos, 7 = 50-54 anos, 8 = 55-59 anos, 9 = 60-64 anos, 10 = 65-69 anos, 11 = 70-74 anos, 12 = 75-79 anos , 13 = 80 anos ou mais)

## 2.4. Ajuste da coluna idade para as analises no dashboards

```python
depara = {
    1: "18-24",
    2: "25-29",
    3: "30-34",
    4: "35-39",
    5: "40-44",
    6: "45-49",
    7: "50-54",
    8: "55-59",
    9: "60-64",
    10: "65-69",
    11: "70-74",
    12: "75-79",
    13: "80+"
}

df["faixa_etaria"] = df["idade"].map(depara)
```

## 2.4. Criando a flag diabetes (variavel alvo)

```python
if 'diabetes' in df.columns:
    df['diabetes_flag']=df['diabetes'].apply(lambda x: 1 if x==2 else 0)
df['diabetes_flag'].value_counts()
```

```python
df.head()
```

## 2.6. Modelo de Pontuação de Risco

```python
# Score de risco

def calc_risco(row):
    score=0
    if 'imc' in row and row['imc']>=30: score+=2
    elif 'imc' in row and row['imc']>=25: score+=1
    if 'pressao_alta' in row and row['pressao_alta']==1: score+=1
    if 'colesterol_alto' in row and row['colesterol_alto']==1: score+=1
    if 'atividade_fisica' in row and row['atividade_fisica']==0: score+=1
    if 'idade' in row and row['idade']>=7: score+=1
    return score

df['risco_total']=df.apply(calc_risco,axis=1)

def cat(x):
    if x>=4: return 'Alto'
    if x>=2: return 'Moderado'
    return 'Baixo'

df['categoria_risco']=df['risco_total'].apply(cat)
df.head()
```

```python
df['categoria_risco'].value_counts()
```
